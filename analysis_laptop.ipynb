## 3. Battery Health by Brand
We group the data to see which laptop brands have the best battery health on average.

```python
# Updated Boxplot to avoid FutureWarning
plt.figure(figsize=(10, 6))

# Assign 'ブランド' to both x and hue, and set legend=False
sns.boxplot(
    data=df, 
    x='ブランド', 
    y='バッテリー健康度（%）', 
    hue='ブランド', 
    palette='Set2', 
    legend=False
)

plt.title('Distribution of Battery Health by Brand')
plt.ylabel('Battery Health (%)')
plt.xlabel('Brand')
plt.show()
```

```python
# Calculate mean battery health per brand
brand_health = df.groupby('ブランド')['バッテリー健康度（%）'].mean().sort_values(ascending=False)

print("Average Battery Health by Brand:")
print(brand_health)

# Visualization: Boxplot of Battery Health by Brand
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='ブランド', y='バッテリー健康度（%）', palette='Set2')
plt.title('Distribution of Battery Health by Brand')
plt.ylabel('Battery Health (%)')
plt.xlabel('Brand')
plt.show()
```

## 4. Correlation Analysis
This code identifies which factors (like age or charge cycles) have the strongest impact on battery health.

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlations = numeric_df.corr()

# Focus on correlations with Battery Health
health_corr = correlations['バッテリー健康度（%）'].sort_values(ascending=False)
print("Correlation with Battery Health:")
print(health_corr)

# Visualization: Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

## 5. Impact of Charge Cycles
Since charge cycles had the strongest negative correlation, we visualize that relationship specifically.

```python
# Scatter plot: Charge Cycles vs. Battery Health
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, 
                x='充電サイクル回数', 
                y='バッテリー健康度（%）', 
                hue='過熱問題の有無',  # Color by Overheating (Yes/No)
                style='ブランド', 
                s=100)

plt.title('Battery Health vs. Charge Cycles')
plt.xlabel('Charge Cycles')
plt.ylabel('Battery Health (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## 6. Performance Ratings Distribution
Finally, we see how users rated their laptop performance.

```python
# Count plot of performance ratings
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='性能評価（レーティング）', palette='viridis')
plt.title('Frequency of Performance Ratings')
plt.xlabel('Rating (1-5)')
plt.ylabel('Count')
plt.show()
```
